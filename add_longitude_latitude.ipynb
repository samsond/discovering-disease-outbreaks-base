```python
from geonamescache import GeonamesCache
from unidecode import unidecode
import pandas as pd
import re
import numpy as np

gc = GeonamesCache()

headline = ''
city = np.nan
latitude = np.nan
longitude = np.nan
countrycode = np.nan
columns = ['headline','city','latitude','longitude','countrycode']
frame = pd.DataFrame(columns=columns)

#TODO 1
#Load in the pandas DataFrame with headline, countries, and cities.
#If a headline contains multiple cities/countries, decide which single one to keep.

data = pd.read_csv('data/extracted_city_country.csv',sep='\t')
#print(data.to_string(index=False))
#print(data["cities"].to_string(index=False))


#TODO 2
#For each city/country, match the name to the latitude and longitude in geonamescache.
#Some cities will return multiple matches with the previous function in different countries. 
#Youâ€™ll have to decide which city to keep based on a heuristic (rule of thumb).
#Add longitude and latitude coordinates to your DataFrame for each headline.
#It will be helpful to get the countrycode of each headline at this point.
#If you were not able to find many countries, think about dropping the column. 
#You also need to decide what to do with headlines that have no coordinates.
#You should end up with over 600 headlines that have geographic coordinates.

for row in data.itertuples(index=True, name='Pandas'):
    matched_cities_by_name = gc.get_cities_by_name(row.cities)
    if matched_cities_by_name:
        cities_selected_by_population = max(matched_cities_by_name,key=lambda x: list(x.values())[0]['population'])
        headline = row.headline.strip('\n')
        latitude = list(cities_selected_by_population.values())[0]['latitude']
        longitude = list(cities_selected_by_population.values())[0]['longitude']
        countrycode = list(cities_selected_by_population.values())[0]['countrycode']
        save_data = {'headline': headline,'city':row.cities,'latitude':latitude,'longitude':longitude,'countrycode':countrycode}
        frame = frame.append(save_data,ignore_index=True)

print(frame.to_string(index=False))

#frame.to_csv('data/extracted_lat_long.csv',index =False, sep='\t')


```
